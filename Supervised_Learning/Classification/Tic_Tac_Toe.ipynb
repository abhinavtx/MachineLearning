```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('tic-tac-toe.csv')
df.head()
```

```python
df.rename(columns={'x': 'top left', 'x.1': 'top middle', 'x.2': 'top right','x.3': 'middle left', 'o': 'middle middle', 'o.1' : 'middle right', 'x.4' : 'bottom left', 'o.2' : 'bottom middle', 'o.3':'bottom right','positive' : 'outcome'},inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
a = df.shape
df.drop_duplicates(inplace=True)
b = df.shape
print(a,b)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y_encod = df.iloc[:,-1]
Target_encod = le.fit_transform(y_encod)
```

```python
col = df.iloc[:,:-1].columns.to_list()
col 
```

```python
X_new = ohe.fit_transform(df[col]).toarray()
newMat = pd.DataFrame(X_new,columns=ohe.get_feature_names_out(col   ))
newMat
```

```python
df = newMat
```

```python
df = pd.concat([df, pd.Series(Target_encod, name='Outcome')], axis=1)
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42, shuffle = True)
```

```python
lr.fit(x_train,y_train)
```

```python
predictedOutcome = lr.predict(x_test)
```

```python
predictedOutcome
```

```python
## Here in classification we use Confusion Matrix for finding accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, predictedOutcome)
print("The confusion matrix is: \n",cm)
print("The accuracy is ",accuracy_score(y_test, predictedOutcome)*100,"%")
```

```python
print(classification_report(y_test, predictedOutcome))
```

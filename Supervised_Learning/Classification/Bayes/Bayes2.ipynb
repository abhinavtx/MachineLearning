```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro


# Function to create distribution plots and conduct Shapiro-Wilk test
def visualize_distribution(dataframe):
    for column in dataframe.iloc[:,:-1].columns:
        # Create a distribution plot (distplot)
        sns.distplot(dataframe[column], hist=True, kde=True, rug=False, bins=30, color='blue', hist_kws={'edgecolor': 'black'})
        plt.title(f'Distribution Plot for {column}')
        plt.xlabel(column)
        plt.ylabel('Density')
        plt.show()

        # Conduct Shapiro-Wilk test and print results
        stat, p_value = shapiro(dataframe[column])
        print(f"Shapiro-Wilk test for {column}: p-value = {p_value}")

# Call the function with your DataFrame
visualize_distribution(df)
```

```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('CarPrice_Assignment.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.nunique()
```

```python
df['cylindernumber'].unique()
```

```python
df.drop(columns=['car_ID','CarName'], inplace= True)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
ohe = OneHotEncoder(drop = 'first')
```

```python
colOhe = ['fueltype','aspiration','doornumber','carbody','drivewheel','enginetype','fuelsystem','enginelocation']
newCol1 = ohe.fit_transform(df[colOhe]).toarray()
newMat = pd.DataFrame(newCol1,columns=ohe.get_feature_names_out(colOhe))
df.drop(columns=colOhe, inplace=True)
newMat.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

order = ['two', 'three', 'four', 'five', 'six', 'eight', 'twelve']

# Initialize the OrdinalEncoder
encoder = OrdinalEncoder(categories=[order])

# Fit and transform the data
encoded_values = encoder.fit_transform(df[['cylindernumber']])

# Replace the original column with encoded values
df['cylindernumber'] = encoded_values

df['cylindernumber']
```

```python
pric = df['price']
df.drop(columns='price', inplace=True)
df = pd.concat([df,newMat],axis=1)
df['price'] = pric
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
```

```python
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42, shuffle = True)
```

```python
z_scores = (df - df.mean()) / df.std()

# Define a threshold for outlier detection (e.g., z-score > 3 or z-score < -3)
threshold = 3

# Find outliers for each column
outliers_mask = (np.abs(z_scores) > threshold)

# Apply capping to outliers in the original DataFrame
df[outliers_mask] = np.sign(df[outliers_mask]) * threshold * df.std() + df.mean()

print("Original DataFrame with outliers capped:")
df 
```

```python
df.shape
```

```python
x_test = sc.transform(x_test)
x_train = sc.transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
linRG = LinearRegression()
linRG.fit(x_train, y_train)
y_pred_LG = linRG.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
treeModel = DecisionTreeRegressor(max_depth=8)
treeModel.fit(x_train, y_train)
y_pred_tree = treeModel.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFmodel = RandomForestRegressor(n_estimators = 300, random_state = 42, min_samples_split=2, min_samples_leaf=1, max_depth=10)
RFmodel.fit(x_train, y_train)
y_pred_RF = RFmodel.predict(x_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
randomSearch = RandomizedSearchCV(RFmodel, param_grid, cv=5, scoring='accuracy')
# Fit the model to the data
randomSearch.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
lrHyperParam = randomSearch.best_params_
print("The hyperparametrs for random forest regression are: ",randomSearch.best_params_)
## This is the final accuracy after cross validation and hyperparamter tuning
print("The r2 score is: ", r2_score(y_test, randomSearch.predict(x_test)))
```

```python
print("R2 score for linear regression is: ", r2_score(y_test, y_pred_LG))
```

```python
print("R2 score for decision tree regression is: ", r2_score(y_test, y_pred_tree))
```

```python
from sklearn.svm import SVR
regres = SVR(kernel='linear')
regres.fit(x_train,y_train)
y_pred_SVR = regres.predict(x_test)
```

```python
print("R2 score for SVM is: ", r2_score(y_test, y_pred_SVR))
```

```python
from sklearn.ensemble import VotingRegressor
voting_regressor = VotingRegressor(estimators=[('tree', treeModel),('forest',RFmodel)])
voting_regressor.fit(x_train, y_train)
```

```python
voting_pred = voting_regressor.predict(x_test)
```

```python
print("R2 score for ensemble regression is: ", r2_score(y_test,voting_pred))
```

```python
## This is the cross validation score
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = RFmodel, X = x_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
## This is the cross validation score
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = voting_regressor, X = x_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
## This is the cross validation score
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = treeModel, X = x_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

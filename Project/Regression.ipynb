```python
# Iterate through each cell in the DataFrame
for index, row in df.loc[:,['registration_year','seats','kms_driven','manufacturing_year','mileage(kmpl)','engine(cc)','max_power(bhp)','torque(Nm)','price(in lakhs)']].iterrows():
    for column in  df.loc[:,['registration_year','seats','kms_driven','manufacturing_year','mileage(kmpl)','engine(cc)','max_power(bhp)','torque(Nm)','price(in lakhs)']].columns:
        # Check if the cell value is not int or float
        if not isinstance(df.at[index, column], (int, float)):
            # Replace the cell value with NaN
            df.at[index, column] = np.nan
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with 7 independent variables and 1 dependent variable
# Specify the features and the target variable
features = df.iloc[:, :-1]  # Select all columns except the last one
target = df.iloc[:, -1]     # Select the last column


# Plot distribution curves for each feature based on the target variable

for i in range(0,(len(features.columns)-1),2):
    plt.figure(figsize=(16,5))
    plt.subplot(1,2,1)
    sns.distplot(df[features.columns[i]])

    plt.subplot(1,2,2)
    sns.distplot(df[features.columns[i+1]])
    #sns.displot(df, x=column, hue=target.name, kind="kde", fill=True)
    plt.show()
    
```

```python
# Calculate the IQR for each feature
Q1 = features.quantile(0.25)
Q3 = features.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers based on the IQR method
outliers = ((features < (Q1 - 1.5 * IQR)) | (features > (Q3 + 1.5 * IQR))).any(axis=1)

# Display the rows with outliers
df_outliers = df[outliers]

df_outliers
```

```python
# Set the lower and upper bounds for capping
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Apply capping to values beyond the bounds
capped_features = features.clip(lower=lower_bound, upper=upper_bound, axis=1)

# Combine capped features and target into a single DataFrame
df_capped = pd.concat([capped_features, target], axis=1)

# Display the DataFrame after capping
df = df_capped
df
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame with multiple independent variables
# Specify the features
features = df.iloc[:, :-1]  # Select all columns except the last one

# Calculate the correlation matrix
correlation_matrix = features.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Create a heatmap for better visualization
plt.figure(figsize=(16, 6))
sns.heatmap(correlation_matrix,  cmap="coolwarm", vmin=-1, vmax=1,annot = True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=42,shuffle=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
x = sc.transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
lrPredicted = lr.predict(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnClass = KNeighborsClassifier(n_neighbors=5,weights='distance',metric='minkowski',algorithm='auto',p=2)
knnClass.fit(x_train,y_train)
knnPredicted = knnClass.predict(x_test)
```

```python
from sklearn.svm import SVC
classifierSVM = SVC(kernel = 'linear', random_state = 0)
classifierSVM.fit(x_train, y_train)
SVMPredicted = classifierSVM.predict(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifierNB = GaussianNB()
classifierNB.fit(x_train, y_train)
NBpredicted = classifierNB.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifierDT = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth = 6, min_samples_split = 5)
classifierDT.fit(x_train, y_train)
DTpredict = classifierDT.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifierRF = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0, max_depth=8, min_samples_split=5)
classifierRF.fit(x_train, y_train)
RFpredict = classifierRF.predict(x_test)
```

```python
NewDf = pd.DataFrame({ "Actual":y_test,'Logistic': lrPredicted, 'KNN':knnPredicted,'SVM':SVMPredicted})
NewDf
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,precision_score,f1_score
cm1 = confusion_matrix(y_test, lrPredicted)
print("The confusion matrix for logistic regression is: \n",cm1)
print("The accuracy of logistic regression is ",accuracy_score(y_test, lrPredicted)*100,"%")
```

```python
## This is the cross validation score
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = lr, X = x_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
cm2 = confusion_matrix(y_test, knnPredicted)
print("The confusion matrix for knn is: \n",cm2)
print("The accuracy of knn is ",accuracy_score(y_test, knnPredicted)*100,"%")
```

```python
cm3 = confusion_matrix(y_test, NBpredicted)
print("The confusion matrix for knn is: \n",cm3)
print("The accuracy of naive bayes is ",accuracy_score(y_test, NBpredicted)*100,"%")
```

```python
## Now using hyperparametr tuning for logistic regression
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  
    'penalty': ['l1', 'l2'],               
    'solver': ['liblinear', 'newton-cg', 'lbfgs', 'sag', 'saga'],  
    'max_iter': [100, 200, 300]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomSearch = RandomizedSearchCV(lr, param_grid, cv=10, scoring='accuracy')

# Fit the model to the data
randomSearch.fit(x_train, y_train)
```

```python
lrHyperParam = randomSearch.best_params_
print("The hyperparametrs for logistic regression are: ",randomSearch.best_params_)
print("The final accuracy score is: ",randomSearch.best_score_*100,"%")
## This is the final accuracy after cross validation and hyperparamter tuning
```
